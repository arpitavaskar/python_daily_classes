```python
# Compile_time/ Method overloading polymorphism
class BCA_1st_semester:
  def student(self,a, b=0,c=0,d=0):
    return a+b+c+d

obj=BCA_1st_semester()
print(obj.student(5))
print(obj.student(10,7,8))
print(obj.student(-10,5,8,-5))
print(obj.student(5,10,15,20))
```

```python
#Run time /Method overring
class Animal:
  def sound(self):
    print("Animal makes a sound")

class Dog(Animal):
  def sound(self):
    print("bow bow bow bow")

class Cat(Animal):
  def sound(self):
    print("meow meow meow meow")

class tiger(Animal):
  def sound(self):
    print("grr grr grr grr")

class snake(Animal):
  def sound(self):
    print("Bussssssssssssssssss")

class Cow(Animal):
  def sound(self):
    print("moo moo moo moo")

obj1=Animal()
obj2=Dog()
obj3=Cat()
obj4=tiger()
obj5=snake()
obj6=Cow()

obj2.sound()
obj3.sound()
obj5.sound()
obj4.sound()
obj6.sound()
```

```python
# Abstraction example
from abc import ABC, abstractmethod
class Game(ABC):
  def Acting(self):
    print("@@@@@@@")

  # NOrmal function / method
  def Free_fire(self):
    print("winner winner chicken dinner")

  #Abstract method
  def BGMI(self):
    pass

 #NOrmal function / method
  def Ludo_king(self):
    print("I'm win the game")

class Temple_run(Game):

  def BGMI(self):
    print(" I got royal pass diamond")

  def Chess(self):
    print("I'm a district player")



obj = Temple_run()
obj.Acting()
obj.Free_fire()
obj.BGMI()
obj.Ludo_king()
obj.Chess()


```
